```python
from perceptron import Perceptron
from sklearn.datasets import load_iris
from random import random, seed
```

We laden 2 maal de iris dataset in aangezien we 2 andere sets nodig hebben

```python
# laat de iris dataset in
sv_iris = load_iris()
vv_iris = load_iris()
```

```python
# set de seed naar mijn leerling nummer
seed(1786733)
```

Voor de opdracht moeten we de Verginica iris buiten de data set laten.

```python
#  hier komen dan dus de Setosa en Versicolour in te zitten
sv_iris["data"] = sv_iris["data"][:100]
sv_iris["target"] = sv_iris["target"][:100]

#  hier komen dan dus de Versicolour en Verginica in te zitten
vv_iris["data"] = vv_iris["data"][50:]
vv_iris["target"] = vv_iris["target"][50:]
```

```python
# maar voor de 2de dataset moeten we de target variablen van 2 naar 0 veraderen
for target in range(len(vv_iris["target"])):
    if vv_iris["target"][target] == 2:
        vv_iris["target"][target] = 0
```

Hier maken we de we de perceptrons aan voor de 2 datasets.

```python
p1 = Perceptron([random(),random(),random(),random()],random())
p2 = Perceptron([random(),random(),random(),random()],random())
```

Deze functie gebruiken we om het gehele train process te automatiseren.

```python
def train_perceptron(perceptron, dataset, epochs)-> None:
    
    print(f"Perceptron Beginning values:\n {perceptron}")

    for epoch in range(epochs):
        print(f"\n=========Epoch {epoch+1}=========")
        count = 0

        for item in dataset["data"]:
            D = dataset["target"][count]
            
            i0 = int(item[0])
            i1 = int(item[1])
            i2 = int(item[2])
            i3 = int(item[3])

            Y = perceptron.activation([i0, i1, i2, i3])

            print(f"Input: {i0, i1, i2, i3} || gekregen Output: {Y} || Correcte Output: {D}")
            perceptron.update(D, Y, [i0, i1], .1)
            
            count+=1

        total_loss = perceptron.loss(4)
        print(f"Total loss: {total_loss}")

        if total_loss == 0 and epoch+1 != epochs:
            print("\nTraining completed early\nStopping process")
            break
    print(f"\nuiteindelijke waarden van de perceptron:\n{perceptron}")
```

Vervolgens gaan we de perceptronen dus nu trainen

```python
train_perceptron(p1, sv_iris, 6)
```

Gelweldig we kunnen nu voor spellen of het gaat om een Setosa of een Versicolour Iris.\
Laten vervolgens de volgende perceptron trainen.

```python
train_perceptron(p2, vv_iris, 20)
```

Het lijkt er op dat dit een zeer lastiger taak is aangezien er nogsteeds een paar kleine fouten gemaakt worden bij het gokken van om welke iris het nou gaat
